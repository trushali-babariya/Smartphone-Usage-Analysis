```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.scatter(
    df['screen_time_hours'],
    df['notifications'],
    c=df['notifications'],
    cmap='viridis',
    alpha=0.7
)

plt.colorbar(label='Number of Notifications')
plt.title("Screen Time vs Notifications")
plt.xlabel("Screen Time (Hours)")
plt.ylabel("Notifications")
plt.show()



```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['day'], df['screen_time_hours'], marker='*',color='orange',              # line color
    linewidth=1,)
plt.title("Daily Screen Time Trend")
plt.xlabel("Day")
plt.ylabel("Screen Time (Hours)")
# plt.legend()
# plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.plot(
    df['day'],
    df['social_media_hours'],
    marker='o',
    linewidth=2,
    label='Social Media Usage',
    color='green'
)

plt.plot(
    df['day'],
    df['gaming_hours'],
    marker='*',
    linewidth=2,
    label='Gaming Usage',
    color='orange'
)

plt.title("Social Media vs Gaming Usage Over Time")
plt.xlabel("Day")
plt.ylabel("Usage Hours")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
activities=["screen time",
            "calls",
            "gaming",
            "social media"]

totals=[df["screen_time_hours"].sum(),
        df["call_duration_hours"].sum(),
        df["gaming_hours"].sum(),
        df["social_media_hours"].sum()]

plt.barh(activities, totals,color='teal')

plt.title("Total Activity Hours Distribution")
plt.xlabel("Total Hours")
plt.ylabel("Activities")
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.show()


```

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("smartphone_usage_365_days 4 1.csv")
plt.figure(figsize=(10, 6))
plt.hist(df['notifications'],bins=15,color='black',edgecolor='k',alpha=0.65)
plt.title("Notification Frequency Distribution")
plt.xlabel("Notification")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df["screen_time_hours"],        # median ko highlight karta hai
    vert=True,           # vertical box plot
    patch_artist=True)
plt.title("Screen Time Distribution")
plt.ylabel("Screen Time (Hours)")
plt.xticks([1], ["Screen Time"])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 7))
values=[df["screen_time_hours"].sum(),
        df["call_duration_hours"].sum(),
        df["gaming_hours"].sum(),
        df["social_media_hours"].sum()]
colors = ['red', 'orange', 'green', 'indigo']
plt.pie(
    values,
    labels=activities,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    labeldistance=1.15,     
    pctdistance=0.7 
)
plt.title("Activity Time Percentage")
plt.axis('equal') 
plt.legend(loc="upper right",bbox_to_anchor=(1.3, 1))  
plt.show()

```
